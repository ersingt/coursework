#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
def my_std(x):
    # Assume x is a numeric iterable

    n = len(x)
    mu = np.average(x)
    return ((1/n) * sum([(val - mu)**2 for val in x]))**0.5
    
# test_x = list(np.random.randint(0,10,7))
# test_x
    
# print(np.std(test_x))
# print(my_std(test_x))
````

````python
def rank_yoy(y1_column_to_rank, y2_column_to_rank, top_ranked=True, rows=10):
    y1_name = list(y1_column_to_rank.columns)[0]
    y2_name = list(y2_column_to_rank.columns)[0]
    
    top_ranked = not top_ranked
    
    if top_ranked:
        rank_order_label = 'Lowest'
    else:
        rank_order_label = 'Highest'
    
    y1 = pd.DataFrame.from_dict(
        {
            state: int(rank + 1) for rank, state in 
            enumerate(y1_column_to_rank.sort_values(y1_name, ascending=top_ranked).head(rows).index)
        }, 
        orient='index')

    y2 = pd.DataFrame.from_dict(
        {
            state: int(rank + 1) for rank, state in 
            enumerate(y2_column_to_rank.sort_values(y2_name, ascending=top_ranked).head(rows).index)
        },
        orient='index')
    
    y1.index.name = 'state'
    y1.columns = ['y1_rank']
    y2.index.name = 'state'
    y2.columns = ['y2_rank']

    y1y2 = y1.merge(y2, how='outer', on='state')
    y1y2.index.name = rank_order_label + ' state'
    
    return y1y2
````

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels, my_bins=10):
    
    nrows = int(np.ceil(len(list_of_columns)/4)) # Makes sure you have enough rows
#     print('nrows = ', nrows)
    fig, ax = plt.subplots(nrows=nrows, ncols=min(4,len(list_of_columns))) # You'll want to specify your figsize
    fig.set_figheight(4)
    fig.set_figwidth(12)
        #https://stackoverflow.com/questions/14770735/how-do-i-change-the-figure-size-with-subplots
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
#         print(i, column)
        ax[i].set(title=list_of_titles[i], xlabel=list_of_xlabels[i])
        ax[i].hist(dataframe[column], bins=my_bins)
````

````python
def subplot_scatterplots(
    dataframe, 
    list_of_x_columns, 
    list_of_y_columns, 
    list_of_titles, 
    list_of_xlabels,
    list_of_ylabels):
    
    nrows = int(np.ceil(len(list_of_x_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2) # You'll want to specify your figsize
    fig.set_figheight(nrows*5)
    fig.set_figwidth(10)
        #https://stackoverflow.com/questions/14770735/how-do-i-change-the-figure-size-with-subplots
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    
    for i, column in enumerate(list_of_x_columns): # Gives us an index value to get into all our lists
        ax[i].set(title=list_of_titles[i], xlabel=list_of_xlabels[i], ylabel=list_of_ylabels[i])
        ax[i].scatter(dataframe[column], dataframe[list_of_y_columns[i]])
````

````python
"""
SAT vs. ACT math scores for 2017
SAT vs. ACT verbal/reading scores for 2017
SAT vs. ACT total/composite scores for 2017
Total scores for SAT 2017 vs. 2018
Composite scores for ACT 2017 vs. 2018
"""

xcol_list = ['sat_2017_math','sat_2017_ebrw','sat_2017_total','sat_2017_total','act_2017_composite']
ycol_list = ['act_2017_math','act_2017_reading','act_2017_composite','sat_2018_total','act_2018_composite']
title_list = [f'{x} vs. {y}' for x, y in zip(xcol_list, ycol_list)]
xlabel_list = [f'{x[:8]}' for x in xcol_list]
ylabel_list = [f'{y[:8]}' for y in ycol_list]

subplot_scatterplots(
    df_final,
    xcol_list,
    ycol_list,
    title_list,
    xlabel_list,
    ylabel_list
)
````
