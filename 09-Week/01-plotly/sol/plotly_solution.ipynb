### Read in sample data that comes in plotly express

```python
df_iris = px.data.iris()
```

```python
df_iris.head()
```

### Make a scatter plot with fanciness

```python
fig = px.scatter(df_iris, x="sepal_width", y="sepal_length", color="species", marginal_y="rug", marginal_x="histogram")
fig
```

### Let's format this code so it complies with style guide. And let's discuss code style.

```python
fig = px.scatter(
    df_iris, 
    x="sepal_width", 
    y="sepal_length", 
    color="species", 
    marginal_y="rug", 
    marginal_x="histogram"
)

fig
```

#### Change something in the plot

```python
my_fig = px.bar(
    df_iris, 
    x="sepal_width", 
    y="sepal_length",
    title='My Chart Title'
)
my_fig
```

```python
                 
my_fig.data[0].update(
    text=df_iris['sepal_length'],  
    textposition=('inside'),
    textfont=dict(size=10),
)
my_fig
```

```python
my_fig.update_xaxes(title='Sepal Width')
```

## cufflinks

use `.iplot()` on a DataFrame

```python
df = pd.DataFrame(np.random.rand(6, 3), columns=['A', 'B', 'C'])
```

```python
df.iplot(kind='bar', barmode='stack', title='Stacked Bar Chart with Random Data')
```

### To modify an attribute of the figure from cufflinks
1. save as a variable and pass `asFigure = True`
2. modify the attribute
3. display figure


```python
my_fig = df.iplot(
    kind='bar', 
    barmode='stack', 
    title='Stacked Bar Chart with Random Data', 
    asFigure=True
)
```

```python
my_fig
```

```python
type(my_fig)
```

#### Setting an attribute

```python
my_fig.layout.yaxis = dict(title='Members', range=[0, 3])
my_fig
```

Or alternatively, use a method. This is more declarative (as opposed to imperative). We don't care what the method does to achieve what we want.

```python
my_fig.update_xaxes(title='Friends')
```

#### Output the interactive file (not just the flat image)

```python
my_fig.write_html("example_plot.html")
```
